```python
# to find the maximum sale 
maxi=df['Weekly_Sales'].max()
maxi
max_sale = df[df['Weekly_Sales'] == maxi]
print("THE MAXIMUM SALE IS ")
max_sale
```

```python
#Which store has maximum standard deviation i.e., the sales vary a lot. Also, find out the coefficient of mean to standard deviation

std_dev = df['Weekly_Sales'].std()
print("Standard Deviation:", std_dev)

mean = df['Weekly_Sales'].mean()
print("Mean:", mean)
cv = std_dev / mean

print("Coefficient of Variation:", cv)
data_series = pd.Series(df['Weekly_Sales'])
max_index = data_series.idxmax()
print("store has maximum standard deviation",max_index)
```

```python
df['Date'] = pd.to_datetime(df['Date'], format='%d-%m-%Y')
df
```

```python
#Which store/s has good quarterly growth rate in Q3’2012
start_date = '2012-01-01'
end_date = '2012-12-31'
filtered_df = df[(df['Date'] >= start_date) & (df['Date'] <= end_date)]
filtered_df
```

```python
#store/s has good quarterly growth rate in Q3’2012
maxi1=filtered_df['Weekly_Sales'].max()
maxi1
max_sale1 = filtered_df[filtered_df['Weekly_Sales'] == maxi1]
print("store/s has good quarterly growth rate in Q3’2012 ")
max_sale1
```

```python
#Change dates into days by creating new variable.
reference_date = df['Date'].min()
df['Day_of_Week'] = df['Date'].dt.day_name()
df['Days'] = (df['Date'] - reference_date).dt.days
print("Change dates into days by creating new variable",df.head())
```

```python
#Some holidays have a negative impact on sales. Find out holidays which have higher sales than the mean sales in non-holiday season for all stores together
non_holiday_sales_mean = df[df['Holiday_Flag'] == False]['Weekly_Sales'].mean()
holiday_sales = df[df['Holiday_Flag'] == True]
high_sales_holidays = holiday_sales[holiday_sales['Weekly_Sales'] > non_holiday_sales_mean]
high_sales_holidays_summary = high_sales_holidays.groupby('Date')['Weekly_Sales'].sum().reset_index()
print("Mean Sales for Non-Holiday Periods:", non_holiday_sales_mean)
print("\nHolidays with Higher Sales than Non-Holiday Mean:")
print(high_sales_holidays_summary)
```

```python
#Provide a monthly and semester view of sales in units and give insights
df['month'] = df['Date']
monthly_sales = df.groupby('month')['Weekly_Sales'].sum().reset_index()
df['semester'] = df['Date'].dt.month.apply(lambda x: 'Semester 1' if x <= 6 else 'Semester 2')
df['year'] = df['Date'].dt.year
semester_sales = df.groupby(['year', 'semester'])['Weekly_Sales'].sum().reset_index()
print("Monthly Sales View:")
print(monthly_sales)
print("\nSemester Sales View:")
print(semester_sales)
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error, r2_score
```

```python

X = df[[ 'CPI', 'Unemployment', 'Fuel_Price']]
Y = df['Weekly_Sales']

X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.3, random_state=42)
model = LinearRegression()
model.fit(X_train, Y_train)
Y_pred = model.predict(X_test)
mse = mean_squared_error(Y_test, Y_pred)
r2 = r2_score(Y_test, Y_pred)
print("THE Linear Regression  ")
print(f'Mean Squared Error: {mse}')
print(f'R-squared: {r2}')
print(f'Intercept: {model.intercept_}')
print(f'Coefficients: {model.coef_}')
```
